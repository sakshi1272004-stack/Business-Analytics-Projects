```python
# Cell 1 - imports and basic settings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score

# Settings
RANDOM_STATE = 42
pd.set_option('display.max_columns', 200)
sns.set_style("whitegrid")
```

```python
# Cell 2 - load data
df = pd.read_csv("C:/Users/227182/OneDrive/Desktop/datasets/Train.csv")

print("Shape:", df.shape)
display(df.head())
print("\nMissing values:")
print(df.isnull().sum())
print("\nColumns:", df.columns.tolist())
```

```python
# Cell 3 - simple missing-value handling (straightforward and interpretable)

# Work on a copy
data = df.copy()

# Numeric columns: fill with median
for col in ["Work_Experience", "Family_Size"]:
    if col in data.columns:
        data[col] = data[col].fillna(data[col].median())

# Categorical columns: fill with mode (most frequent)
for col in ["Gender", "Ever_Married", "Graduated", "Profession", "Spending_Score", "Var_1"]:
    if col in data.columns:
        data[col] = data[col].fillna(data[col].mode()[0])

# Confirm
print("After imputation — missing values:")
print(data.isnull().sum())
```

```python
# Cell 4 - choose features for clustering (demographic + spending behaviour)
features = ["Gender", "Ever_Married", "Age", "Graduated",
            "Profession", "Work_Experience", "Spending_Score", "Family_Size", "Var_1"]

X = data[features].copy()

# Convert categorical variables using get_dummies (simple & easy to interpret)
X_encoded = pd.get_dummies(X, drop_first=True)

print("Feature matrix shape (after one-hot):", X_encoded.shape)
display(X_encoded.head())
```

```python
# Cell 5 - scale numeric columns so KMeans treats each feature equally
scaler = StandardScaler()

# We scale the whole encoded matrix (it's fine because one-hot columns are 0/1)
X_scaled = scaler.fit_transform(X_encoded)

# Quick check
print("Scaled matrix shape:", X_scaled.shape)
```

```python
# Cell 6 - elbow and silhouette method (k from 2 to 8)
wcss = []
sil_scores = []
K_range = range(2,9)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=RANDOM_STATE, n_init=10)
    labels = kmeans.fit_predict(X_scaled)
    wcss.append(kmeans.inertia_)
    sil_scores.append(silhouette_score(X_scaled, labels))

# Plot results
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(list(K_range), wcss, marker='o')
plt.title("Elbow method (WCSS)")
plt.xlabel("k")
plt.ylabel("WCSS")

plt.subplot(1,2,2)
plt.plot(list(K_range), sil_scores, marker='o')
plt.title("Silhouette scores")
plt.xlabel("k")
plt.ylabel("Silhouette")

plt.tight_layout()
plt.show()

# Print values so you can choose
for k, w, s in zip(K_range, wcss, sil_scores):
    print(f"k={k} → WCSS={w:.0f} , Silhouette={s:.3f}")

# AUTO-SELECT by silhouette (you can override this)
best_k = K_range[int(np.argmax(sil_scores))]
print("\nAuto-chosen k (best silhouette):", best_k)
```

```python
# Cell 7 - fit KMeans with chosen k
chosen_k = best_k   # change this number manually if you prefer another k after inspecting plots
kmeans = KMeans(n_clusters=chosen_k, random_state=RANDOM_STATE, n_init=20)
labels = kmeans.fit_predict(X_scaled)
data["Cluster"] = labels

print("Cluster counts:")
print(data["Cluster"].value_counts().sort_index())
```

```python
# Cell 8 - PCA projection for visualization
pca = PCA(n_components=2, random_state=RANDOM_STATE)
proj = pca.fit_transform(X_scaled)

plt.figure(figsize=(8,6))
palette = sns.color_palette("tab10", n_colors=chosen_k)
for cl in sorted(data["Cluster"].unique()):
    mask = data["Cluster"] == cl
    plt.scatter(proj[mask,0], proj[mask,1], label=f"Cluster {cl}", alpha=0.6, s=25)
plt.xlabel("PC1"); plt.ylabel("PC2")
plt.title(f"PCA 2D view of clusters (k={chosen_k})")
plt.legend()
plt.show()
```

```python
# Cell 9 - profile each cluster with numeric means and top categorical values

# Numeric summaries
num_summary = data.groupby("Cluster")[["Age", "Work_Experience", "Family_Size"]].agg(["count","mean","median"]).round(2)
print("Numeric summary per cluster:")
display(num_summary)

# Top categories per cluster (as percentages)
cat_cols = ["Gender", "Ever_Married", "Graduated", "Profession", "Spending_Score", "Var_1"]
for col in cat_cols:
    print(f"\nTop values of {col} per cluster (percent):")
    pct = data.groupby("Cluster")[col].value_counts(normalize=True).mul(100).round(1)
    pct_df = pct.reset_index(name="pct").sort_values(["Cluster","pct"], ascending=[True, False])
    display(pct_df.groupby("Cluster").head(5))

# Aggregated one-line summary table
agg = data.groupby("Cluster").agg(
    customers = ("ID","count"),
    age_mean = ("Age","mean"),
    famsize_mean = ("Family_Size","mean")
).round(2)
agg["top_spending_score"] = data.groupby("Cluster")["Spending_Score"].agg(lambda x: x.value_counts().idxmax())
print("\nAggregated cluster table:")
display(agg)
```
