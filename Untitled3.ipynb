```python
# Drop completely empty columns (if any)
df = df.dropna(axis=1, how='all')

# Fill missing values (example for Salary)
if 'MonthlyIncome' in df.columns:
    df['MonthlyIncome'].fillna(df['MonthlyIncome'].median(), inplace=True)
```

```python
financial_cols = [c for c in df.columns if c.lower() in
                  ["monthlyincome", "hourlyrate", "dailyrate",
                   "percentsalaryhike", "bonus", "salary"]]
print("\nFinancial Columns found:", financial_cols)
```

```python
# Convert to numeric and clean
for col in financial_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df[financial_cols] = df[financial_cols].fillna(df[financial_cols].median())
```

```python
# Analysis & Visuals
# Average income by department
if set(["Department","MonthlyIncome"]).issubset(df.columns):
    avg_income = df.groupby("Department")["MonthlyIncome"].mean().sort_values()
    plt.figure(figsize=(8,4))
    avg_income.plot(kind="bar", color="steelblue")
    plt.title("Average Monthly Income by Department")
    plt.ylabel("Income")
    plt.tight_layout()
    plt.show()
```

```python
# Income distribution
if "MonthlyIncome" in df.columns:
    plt.figure(figsize=(7,4))
    sns.histplot(df["MonthlyIncome"], kde=True, color="green")
    plt.title("Monthly Income Distribution")
    plt.xlabel("Monthly Income")
    plt.tight_layout()
    plt.show()
```

```python

# Attrition vs Income
if set(["Attrition","MonthlyIncome"]).issubset(df.columns):
    plt.figure(figsize=(6,4))
    sns.boxplot(x="Attrition", y="MonthlyIncome", data=df)
    plt.title("Income by Attrition Status")
    plt.tight_layout()
    plt.show()
```

```python
# Correlation Heatmap for financial columns
if financial_cols:
    corr_data = df[financial_cols].apply(pd.to_numeric, errors='coerce')
    corr = corr_data.corr()
    if not corr.empty:
        plt.figure(figsize=(6,5))
        sns.heatmap(corr, annot=True, cmap="Blues")
        plt.title("Correlation among Financial Metrics")
        plt.tight_layout()
        plt.show()
    else:
        print("No numeric financial columns for correlation.")
else:
    print("No financial columns found.")
```
