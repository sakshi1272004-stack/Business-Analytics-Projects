```python
import pandas as pd # import the libraries
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# Load CSV with fallback encoding
df = pd.read_csv("C:/Users/227182/OneDrive/Desktop/datasets/sales_data_sample.csv", encoding="latin1")
```

```python
df
```

```python
# --- Convert Dates ---
df["ORDERDATE"] = pd.to_datetime(df["ORDERDATE"], errors="coerce")

# --- Remove Duplicates ---
df = df.drop_duplicates()

# --- Strip Whitespace in all text columns ---
str_cols = df.select_dtypes(include="object").columns
df[str_cols] = df[str_cols].apply(lambda x: x.str.strip())

# --- Standardize Column Names ---
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# --- Check and Handle Missing Values ---
print("\nMissing values before cleaning:\n", df.isna().sum())

# Drop addressline2 if present
if "addressline2" in df.columns:
    df = df.drop(columns=["addressline2"])

# Fill missing in state & territory with their mode
for col in ["state", "territory"]:
    if col in df.columns:
        mode_val = df[col].mode().iloc[0] if not df[col].mode().empty else None
        if mode_val:
            df[col] = df[col].fillna(mode_val)

# Fill postalcode using forward fill
if "postalcode" in df.columns:
    df["postalcode"] = df["postalcode"].fillna(method="ffill")

print("\nMissing values after cleaning:\n", df.isna().sum())

# --- Encode Categorical Variable (dealsize) ---
if "dealsize" in df.columns:
    le = LabelEncoder()
    df["dealsize_encoded"] = le.fit_transform(df["dealsize"])

# --- Remove Outliers in sales using IQR ---
if "sales" in df.columns:
    Q1 = df["sales"].quantile(0.25)
    Q3 = df["sales"].quantile(0.75)
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
    df = df[(df["sales"] >= lower) & (df["sales"] <= upper)]
```

```python
df.to_csv("sales_data_cleaned.csv", index=False)
print("✅ Cleaned dataset saved as sales_data_cleaned.csv")

print("\nCleaned Dataset Info:")
df.info()
```

```python
# Extract Year, Month, Quarter
df["year"] = df["orderdate"].dt.year
df["month"] = df["orderdate"].dt.month
df["quarter"] = df["orderdate"].dt.quarter

# Sales per Unit
df["sales_per_unit"] = df["sales"] / df["quantityordered"]
```

```python
# Total sales by Year
sales_year = df.groupby("year")["sales"].sum().reset_index()

# Sales by Month (all years combined)
sales_month = df.groupby("month")["sales"].sum().reset_index()

# Sales by Product Line
sales_product = df.groupby("productline")["sales"].sum().sort_values(ascending=False)

# Country-wise sales
sales_country = df.groupby("country")["sales"].sum().sort_values(ascending=False)

print(sales_year.head())
print(sales_month.head())
print(sales_product.head())
print(sales_country.head())
```

```python
# --- Yearly Sales Trend ---
plt.figure(figsize=(8,5))
plt.plot(sales_year["year"], sales_year["sales"], marker="o")
plt.title("Total Sales by Year")
plt.xlabel("Year")
plt.ylabel("Sales ($)")
plt.grid(True)
plt.show()
```

```python
# --- Monthly Sales Trend ---
plt.figure(figsize=(10,5))
plt.bar(sales_month["month"], sales_month["sales"], color="teal")
plt.title("Total Sales by Month")
plt.xlabel("Month")
plt.ylabel("Sales ($)")
plt.show()
```

```python
# --- Sales by Product Line ---
plt.figure(figsize=(10,5))
sales_product.plot(kind="bar", color="orange")
plt.title("Sales by Product Line")
plt.ylabel("Sales ($)")
plt.show()
```

```python
# --- Country-wise Sales (Top 10) ---
plt.figure(figsize=(10,5))
sales_country.head(10).plot(kind="bar", color="steelblue")
plt.title("Top 10 Countries by Sales")
plt.ylabel("Sales ($)")
plt.show()
```
